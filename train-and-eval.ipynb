```
# Download the falko-marlin parallel corpus
!curl -v http://www.sfs.uni-tuebingen.de/~adriane/download/wnut2018/data.tar.gz \
        --output gector/data/falko-marlin.tar.gz && mkdir gector/data/falko-marlin/ \
        && tar -xzf gector/data/falko-marlin.tar.gz -C gector/data/falko-marlin/
```

```
# Preprocess training and dev sets into format for gector model
!python gector/utils/preprocess_data.py -s gector/data/falko-marlin/data/fm-train.src \
        -t gector/data/falko-marlin/data/fm-train.trg \
        -o gector/data/fm.train
```

```
!python gector/utils/preprocess_data.py -s gector/data/falko-marlin/data/fm-dev.src \
        -t gector/data/falko-marlin/data/fm-dev.trg \
        -o gector/data/fm.dev
```

```
# Train!
!python gector/train.py --train_set gector/data/fm.train \
    --dev_set gector/data/fm.dev \
    --model_dir gector/model/ \
    --transformer_model bert_de
```

```
!python gector/predict.py --model_path gector/model/model.th \
    --vocab_path gector/model/vocabulary/ \
    --transformer_model bert_de \
    --input_file gector/data/falko-marlin/data/fm-test.src \
    --output_file gector/data/fm.preds \
    --mode file
```
